```python
jobs_df = jobs_df.loc[jobs_df['Job Title'].str.contains('Data Scientist')]
```

```python
jobs_df['Salary Estimate'].value_counts()
```

```python
jobs_df['Salary Estimate'] = jobs_df['Salary Estimate'].str.replace('(Employer est.)', ' (Employer est.)', regex=False)
```

```python
jobs_df['Salary Estimate'].value_counts()
```

```python
jobs_df['Salary Estimate'] = jobs_df['Salary Estimate'].apply(lambda x: x.split('-')).apply(lambda x: (x[0], x[1].split()[0]))
```

```python
jobs_df['Salary Estimate'] = jobs_df['Salary Estimate'].apply(lambda x: (int(x[0][1:-1]),int(x[1][1:-1])))
```

```python
jobs_df['Salary Estimate'] = jobs_df['Salary Estimate'].apply(lambda x: (x[0]+x[1]) / 2 )
```
