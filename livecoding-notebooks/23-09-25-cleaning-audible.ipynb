```python
# Remove extra characters from author and narrator columns
audible['author'] = audible.author.str.split(':').apply(lambda x: x[-1])
```

```python
audible
```

```python
audible['narrator'] = audible.narrator.str.split(':').apply(lambda x: x[-1])
```

```python
audible
```

```python
# How often is the author the same person as the narrator? 

audible['read_by_author'] = audible.author == audible.narrator
```

```python
# Convert time into a single number 


def lst_to_time(lst):
    if len(lst) == 0:
        return 0.0
    
    hrs = float(lst[0])
    if len(lst) == 1:
        return hrs
    
    return hrs + (float(lst[1]) / 60)




audible['runtime'] = (audible.time.str
                     .split()
                     .apply(lambda lst: [item for item in lst if item.isnumeric()])
                     .apply(lst_to_time))
```

```python
audible.sample(10)
```

```python
audible = audible.drop(columns=['time'])
```

```python
# Release date format: dd/mm/yy
audible.releasedate = pd.to_datetime(audible.releasedate, format='%d-%m-%y')
```

```python
# Stars should be two separate columns
audible.stars
```

```python
# Use a pre-defined function to transform the split stars column

def stars_to_num_ratings(lst):
    if len(lst) == 3:
        return 0
    substr = lst[4][5:]
    substr = substr.replace(',', '')
    return int(substr)

audible['score'] = (
    audible.stars.str
    .split()
    .apply(lambda lst: 0.0 if len(lst) == 3 else float(lst[0]))
)



audible['num_ratings'] = (
    audible.stars.str
    .split()
    .apply(stars_to_num_ratings)
)
```

```python
audible.sample(10)
```

```python
audible = audible.drop(columns=['stars'])
```

```python
audible
```

```python
# How can we group on day and month, irrespective of year? 

audible['month'] = audible.releasedate.dt.month
audible['day'] = audible.releasedate.dt.day
audible['monthday'] = audible.releasedate.dt.month + (audible.releasedate.dt.day/100)
```

```python
audible.groupby(['month', 'day']).size().reset_index()
```

```python
monthday_grouping = audible.groupby('monthday').size()
```

```python
# View where errors occur in the conversion of the index to date-time 
monthday_grouping[pd.to_datetime(monthday_grouping.index.astype(str), format='%m.%d', errors='coerce').isna()]
```
